```python
sns.set()

# TASK OUTPUT PATHS
task_1_output = '../mongo/outputs/task_1/task_1.csv'
task_2_output = '../mongo/outputs/task_2/task_2.csv'
task_3_output = '../mongo/outputs/task_3/task_3.csv'
task_4_output = '../mongo/outputs/task_4/task_4.csv'
task_5_output = '../mongo/outputs/task_5/task_5.csv'
```

```python
df_task_1 = pd.read_csv(task_1_output)

plt.figure(figsize=(8,5))
ax = sns.countplot(x="type", data=df_task_1, hue="type", dodge=False)
plt.title("Movie and TV Show productions in Netflix for 2019")
plt.xlabel("Type")
plt.ylabel("Productions")
plt.legend()
plt.show()
```

```python
df_task_2 = pd.read_csv(task_2_output)
sns.set(font_scale=2.3)
plt.figure(figsize=(35,15))
ax = sns.barplot(x = '_id', y = 'count', data = df_task_2)
plt.title("Productions per country")
plt.xlabel("Country")
plt.ylabel("Productions")
plt.xticks(rotation=90)
plt.show()
```

```python
df_task_3 = pd.read_csv(task_3_output)
# df_task_3
sns.set(font_scale=2.3)
plt.figure(figsize=(35,15))
ax = sns.barplot(x = '_id', y = 'count', data = df_task_3)
plt.title("Productions per genre")
plt.xlabel("Genre")
plt.ylabel("Productions")
plt.xticks(rotation=90)
plt.show()
```

```python
df_task_4 = pd.read_csv(task_4_output)
# df_task_4[:20]
sns.set(font_scale=2.3)
plt.figure(figsize=(35,15))
ax = sns.barplot(x = '_id', y = 'count', data = df_task_4[:20])
plt.title("Top 20 participating actors")
plt.xlabel("Actors")
plt.ylabel("Productions")
plt.xticks(rotation=90)
plt.show()
```

```python
df_task_5 = pd.read_csv(task_5_output)
df_task_5 = df_task_5.sort_values(by=['count'], ascending = False)
actors_num = 30
sns.set(font_scale=2.3)
plt.figure(figsize=(35,15))

ax = sns.barplot(x = '_id', y = 'count', data = df_task_5[:actors_num])
i = 0
for p in ax.patches:
    ax.annotate(format(df_task_5.loc[i,'type'], 's'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()/2), 
                   ha = 'center', va = 'center', rotation=90, color = 'white',
                   xytext = (0, 9), 
                   textcoords = 'offset points')
    i += 1
plt.title("Top 20 appearing actors in their genre of preference")
plt.xlabel("Actors")
plt.ylabel("Productions")
plt.xticks(rotation=90)
plt.show()
```
